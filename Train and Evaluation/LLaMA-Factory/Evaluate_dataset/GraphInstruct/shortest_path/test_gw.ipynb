```python
error_cnt = 0
correct_cnt = 0
for i, data in enumerate(datas):
    text = data['code']
    ans = gt[i]['answer']


    # 获取输出内容
    output_content = text
    output_content = str(output_content).strip()
    # print("Captured Output:")
    # print(output_content)
    
    if '### Response' in output_content:
        output_content = output_content.split('### Response')[1].split('###')[1].split('.')[0].strip()
    
    elif '###' in output_content:
        output_content = output_content.split('###')[1].split('.')[0].strip()

    
    
    # print(f'llm={output_content} gt={ans}')
        
    if output_content.strip() == str(ans).strip():
        correct_cnt = correct_cnt + 1
        # print(gt[int(i)]['input_prompt'])
    else:
        # print(gt[int(i)]['input_prompt'])
        print(output_content,ans)
        # print(code)
```
